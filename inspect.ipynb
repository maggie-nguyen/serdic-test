```python
from ultralytics import YOLO

# Load your trained model
model = YOLO('models/20260324_human.pt')

# 1. Access a comprehensive summary of the model
model.info()

# 2. Access class names (labels)
print("\nClass Names:")
print(model.names)

# 3. Access training/model configuration (like image size)
print("\nModel Configuration (Overrides):")
print(model.overrides)

# Example: Access specific attributes directly
img_sz = model.overrides.get('imgsz', 'Not specified')
print(f"\nTraining Image Size: {img_sz}")
print(f"Task type: {model.task}")
```
